```python
# Beat tracking example
import librosa
```

```python
# 1. Get the file path to an included audio example

#file path of the audio file
filename = librosa.example('nutcracker')
```

```python
type(filename)
```

### 2. Load the audio as a waveform `y`


```python
#    Store the sampling rate as `sr`

# y as one dimensional numpy array in time series, corresponds to amplitude of the waveform at sample t

# sr as positive integer, sampling rate, number of samples per second
# default is all audio mixed to "mono" and resampled to 22050 Hz when loaded, can specificy in librosa.load argument

y, sr = librosa.load(filename)
```

```python
y # audio pitch values array
```

```python
len(y)
```

```python
sr # sampling rate, 22050 hz, or 22050 samples per second
```

### 3. Run the default beat tracker

```python
# two outputs from librosa beat tracker

# tempo saving estimate tempo (in beats per minute)
# beat_frame saving array of frame numbers with detected beat events

# frames as windows of the signal y, each seperated by hop_length = 512 samples
#     slice of time series, usually equivalent to single column of a spectogram matrix

tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
```

```python
tempo # rate of beats per minute
```

```python
beat_frames # frames with a beat?
```

```python
print('Estimated tempo: {:.2f} beats per minute'.format(tempo))
```

```python
print(f'Estimated tempo: {tempo:.2f} beats per minute')
```

### 4. Convert the frame indices of beat events into timestamps

```python
# converting frame numbers "beat_frames" into timings, 
#     so array of timestamps in seconds at times of detected beat events
beat_times = librosa.frames_to_time(beat_frames, sr=sr)
```

```python
sr = 22050

y_sweep = librosa.chirp(fmin=librosa.note_to_hz('C3'),
                        fmax=librosa.note_to_hz('C5'),
                        sr=sr,
                        duration=1)

Audio(data=y_sweep, rate=sr)
```

```python
import matplotlib.pyplot as plt
y, sr = librosa.load(librosa.ex('choice'), duration=10)
fig, ax = plt.subplots(nrows=3, sharex=True)
librosa.display.waveshow(y, sr=sr, ax=ax[0])
ax[0].set(title='Envelope view, mono')
ax[0].label_outer()

y, sr = librosa.load(librosa.ex('choice', hq=True), mono=False, duration=10)
librosa.display.waveshow(y, sr=sr, ax=ax[1])
ax[1].set(title='Envelope view, stereo')
ax[1].label_outer()


y, sr = librosa.load(librosa.ex('choice'), duration=10)
y_harm, y_perc = librosa.effects.hpss(y)
librosa.display.waveshow(y_harm, sr=sr, alpha=0.5, ax=ax[2], label='Harmonic')
librosa.display.waveshow(y_perc, sr=sr, color='r', alpha=0.5, ax=ax[2], label='Percussive')
ax[2].set(title='Multiple waveforms')
ax[2].legend();
```

```python
fig, (ax, ax2) = plt.subplots(nrows=2, sharex=True)
ax.set(xlim=[6.0, 6.01], title='Sample view', ylim=[-0.2, 0.2])
librosa.display.waveshow(y, sr=sr, ax=ax, marker='.', label='Full signal')
librosa.display.waveshow(y_harm, sr=sr, alpha=0.5, ax=ax2, label='Harmonic')
librosa.display.waveshow(y_perc, sr=sr, color='r', alpha=0.5, ax=ax2, label='Percussive')
ax.label_outer()
ax.legend()
ax2.legend();
```

```python
fig, ax = plt.subplot_mosaic("hSSS;hSSS;hSSS;.vvv")
y, sr = librosa.load(librosa.ex('trumpet'))
chroma = librosa.feature.chroma_cqt(y=y, sr=sr)
sim = librosa.segment.recurrence_matrix(chroma, mode='affinity')
librosa.display.specshow(sim, ax=ax['S'], sr=sr,
                         x_axis='time', y_axis='time',
                         auto_aspect=False)
ax['S'].label_outer()
ax['S'].sharex(ax['v'])
ax['S'].sharey(ax['h'])
ax['S'].set(title='Self-similarity')
librosa.display.waveshow(y, ax=ax['v'])
ax['v'].label_outer()
ax['v'].set(title='transpose=False')
librosa.display.waveshow(y, ax=ax['h'], transpose=True)
ax['h'].label_outer()
ax['h'].set(title='transpose=True');
```
